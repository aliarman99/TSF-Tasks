### Importing Libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
from sklearn import tree
```

### Data reading

```python

# Load data and store it into pandas DataFrame objects
iris = load_iris()
X = pd.DataFrame(iris.data[:, :], columns = iris.feature_names[:])
y = pd.DataFrame(iris.target, columns =["Species"])
```

### Split Data

```python
# Defining and fitting a DecisionTreeClassifier instance

from sklearn.model_selection import train_test_split # Import train_test_split function
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1) # 60% training and 40% test
```

### Training Phase

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)

print("Completed")
```

### Visualising 

```python
fig = plt.figure(figsize=(25,20))
tree1 = tree.plot_tree(clf,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
```

### Model Evaluation

```python
from sklearn import metrics 
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
